```python
import pandas as pd
import numpy as np
```

```python
cafe_Df=pd.read_csv("../data/dirty_cafe_sales.csv")
```

```python
cafe_Df.head()
```

```python
cafe_Df.isna().sum()
```

```python
for i in cafe_Df.columns:
    null_rate= cafe_Df[i].isnull().sum()/len(cafe_Df)*100
    if null_rate >0:
        print( f'{i} null rate: {null_rate.round(2)}%')
```

```python
cafe_Df[["Quantity","Price Per Unit","Total Spent"]]=cafe_Df[["Quantity","Price Per Unit","Total Spent"]].apply(pd.to_numeric, errors="coerce")
cafe_Df[:5]
```

```python
print(cafe_Df["Total Spent"].unique())
```

```python
cafe_Df[["Price Per Unit","Quantity"]].loc[1.0:"Nan"]
```

```python
cafe_Df["Price Per Unit"] = cafe_Df["Price Per Unit"].fillna(cafe_Df["Total Spent"] /cafe_Df["Quantity"])
cafe_Df.isna().sum()
```

```python
cafe_Df["Quantity"]= cafe_Df["Quantity"].fillna(cafe_Df["Total Spent"] /cafe_Df["Price Per Unit"])
```

```python
cafe_Df["Total Spent"]= cafe_Df["Total Spent"].fillna(cafe_Df["Price Per Unit"]* cafe_Df["Quantity"])
cafe_Df.isna().sum()
```

```python
cafe_Df["Transaction Date"]=pd.to_datetime(cafe_Df["Transaction Date"],errors="coerce")
cafe_Df.head()
```

```python
cafe_Df.replace(['ERROR','UNKNOWN'],np.nan, inplace=True)
```

```python
cafe_Df['Transaction Date']= pd.to_datetime(cafe_Df['Transaction Date'],errors='coerce')
```

```python
median_Date= cafe_Df['Transaction Date'].mean()
cafe_Df['Transaction Date']= cafe_Df['Transaction Date'].fillna(median_Date)
```

```python
cat_col= ['Payment Method', 'Location']
for col in cat_col:
  cafe_Df[col]=cafe_Df[col].fillna(cafe_Df[col].mode()[0])
```

```python
cafe_Df= cafe_Df.fillna(cafe_Df.mean(numeric_only=True))
cafe_Df.isna().sum()
```

```python
cols_to_round = ['Price Per Unit', 'Quantity', 'Total Spent']
def round_to_half(x):
    return round(x *2) /2
for cols in cols_to_round:
    cafe_Df[cols]= cafe_Df[cols].apply(round_to_half)
```

```python
cafe_Df['Price Per Unit'].unique()
```

```python
price_to_item = {
    2.0: 'Coffee',
    1.5: 'Tea',
    4.0: 'Sandwich',
    5.0: 'Salad',
    3.0: 'Cake',
    1.0: 'Cookie',
}
```

```python
mask=  cafe_Df['Item'].isna() & cafe_Df['Price Per Unit'].notna()
cafe_Df.loc[mask,'Item'] = cafe_Df.loc[mask, 'Price Per Unit'].map(price_to_item)
```

```python
cafe_Df.isna().sum()
```
