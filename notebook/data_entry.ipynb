```python
df_prof['Country']= df_prof['Country'].str.lower()
df_prof['Country']= df_prof['Country'].str.strip()
```

```python
matches = fuzzywuzzy.process.extract("south korea",countires, limit=10,scorer=fuzzywuzzy.fuzz.token_sort_ratio)
matches
```

```python
def replace_matches_column(df, columm, string_to_match, min_ratio= 47):
    strings = df[columm].unique()
    matches= fuzzywuzzy.process.extract(string_to_match, strings, limit=10, scorer=fuzzywuzzy.fuzz.token_sort_ratio)

    close_matches=[matches[0] for matches in matches if matches [1] >= min_ratio]

    row_with_matches= df[columm].isin(close_matches)
    df.loc[row_with_matches, columm]= string_to_match

    print("All done")
```

```python
replace_matches_column(df= df_prof ,columm='Country', string_to_match='South korea')
```
